```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import mplcursors  # Import mplcursors

# Read data from your CSV file (replace with your actual file path)
csv_file = "csv/neo.csv"
df = pd.read_csv(csv_file)

# Create a scatter plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    x='Closest approach date',
    y='How much did the asteroid miss us by? (km)',
    hue='Est diameter (km)',
    size='Est diameter (km)',
    data=df,
    palette='viridis',
    legend='brief',
)

# Initialize mplcursors for hover and click behavior
cursor = mplcursors.cursor(hover=True, multiple=True)

# Update annotation text when hovering over a point
@cursor.connect("add")
def on_add(sel):
    ind = sel.target.index
    text = f"{df['Asteroid ID'][ind]} ({df['Name'][ind]})"
    sel.annotation.set_text(text)

# Rotate x-axis labels
plt.xticks(rotation=90)
plt.xlabel('Closest Approach Date')
plt.ylabel('Distance from Earth (km)')
plt.title('Asteroid Miss Distance vs. Diameter')
plt.grid(True)
plt.legend(title='Potentially Hazardous?')

# Move the legend outside to the right
plt.legend(title='Potentially Hazardous?', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
# calc abs distance
df[('distance')] = df['How much did the asteroid miss us by? (km)']
```

```python
# create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['distance'], [0] * len(df), c='b', label='Close Approach Asteroid Data')
plt.scatter(0, 0, c='r', marker='o', s=100, label='Center Point of Earth')
plt.xlabel('Distance from Center (km)')
plt.ylabel('Y')
plt.title('Relative Distance from Earth Center')
plt.grid(True)
plt.show()
```

```python
# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(df['Closest approach date'], df['How much did the asteroid miss us by? (km)'], marker='+', linestyle='-', label='Asteroids', markersize=df['Est diameter (km)'])

plt.xlabel('Closest Approach Date')
plt.ylabel('Distance from Earth (km)')
plt.title('Asteroid Miss Distance vs. Diameter')
plt.grid(True)
plt.legend()
plt.show()
```
